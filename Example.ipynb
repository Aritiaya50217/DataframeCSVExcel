```python
# check 
data['Store'] == 1
```

```python
# นับว่า Store == 1 มีเท่าไร

st1 = data[data['Store'] == 1]
st1
```

```python
# Add attributes to data
st1['Unemployment'].describe()
```

```python
# สร้างฟังก์ชัน
def label_unemploy(rate):
    # if  Unemployment >= 8 : High
    if rate >= 8 :
        return 'High'
    #  elif  Unemployment >= 7 : Medium
    elif rate >= 7 :
        return 'Medium'
    #  else : Low
    else:
        return 'Low'


label_unemploy(8) 
```

```python
# apply ฟังก์ชัน
# เพิ่ม column  'Unemploy_label' ท้ายสุด
st1['Unemploy_label'] = st1['Unemployment'].apply(label_unemploy)
```

```python
# แสดงค่า object in function
st1['Unemploy_label'].unique()
```

```python
# ดูข้อมูลใน column Unemployment
st1['Unemployment'].unique()
```

```python
# แปลงเป็นจำนวนเต็ม
st1['Unemployment'] = st1['Unemployment'] /100
```
